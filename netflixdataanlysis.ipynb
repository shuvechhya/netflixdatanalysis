```python
# Extract numerical values from the 'duration' column
netflix_df['duration'] = netflix_df['duration'].str.extract('(\d+)').astype(float)

# Create separate DataFrames for Movies and TV Shows
movies_df = netflix_df[netflix_df['type'] == 'Movie']
tv_shows_df = netflix_df[netflix_df['type'] == 'TV Show']

# Plotting histograms for the distribution of durations
plt.figure(figsize=(12, 6))
sns.histplot(movies_df['duration'], bins=30, kde=True, label='Movies', color='skyblue')
sns.histplot(tv_shows_df['duration'], bins=30, kde=True, label='TV Shows', color='orange')
plt.title('Distribution of Durations for Movies and TV Shows')
plt.xlabel('Duration (min/seasons)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Movies tend to have a diverse range of durations, with a concentration around 90 to 120 minutes.
TV shows vary in duration, with 1 to 3 seasons being common.

```python
netflix_df['genres'] = netflix_df['listed_in'].str.split(', ')

movies_df = netflix_df[netflix_df['type'] == 'Movie']
tv_shows_df = netflix_df[netflix_df['type'] == 'TV Show']

# Flatten the list of genres for Movies and TV Shows
all_movie_genres = [genre for genres_list in movies_df['genres'].dropna() for genre in genres_list]
all_tv_show_genres = [genre for genres_list in tv_shows_df['genres'].dropna() for genre in genres_list]

# Plotting countplots for the distribution of genres
plt.figure(figsize=(14, 8))

# Countplot for Movie Genres
plt.subplot(2, 1, 1)
sns.countplot(y=all_movie_genres, order=pd.Series(all_movie_genres).value_counts().index, palette='viridis')
plt.title('Distribution of Genres in Movies')
plt.xlabel('Count')
plt.ylabel('Genre')

# Countplot for TV Show Genres
plt.subplot(2, 1, 2)
sns.countplot(y=all_tv_show_genres, order=pd.Series(all_tv_show_genres).value_counts().index, palette='muted')
plt.title('Distribution of Genres in TV Shows')
plt.xlabel('Count')
plt.ylabel('Genre')

plt.tight_layout()
plt.show()
```

Netflix offers a diverse range of genres in both movies and TV shows, catering to a wide audience.

Dramas, Comedies, and International content are consistently popular across both movies and TV shows.

The popularity of specific genres can provide insights into audience preferences and help shape content acquisition and creation strategies.

```python
# Group by director and count the number of movies/TV shows directed by each
director_counts = netflix_df.groupby('director')['show_id'].count().reset_index()

director_counts = director_counts.sort_values(by='show_id', ascending=False)

top_directors = director_counts.head(10) 
print(top_directors)
```

```python
plt.figure(figsize=(6, 6))
plt.bar(top_directors['director'], top_directors['show_id'], color='skyblue')
plt.xlabel('Director')
plt.ylabel('Number of Shows Directed')
plt.title('Top Directors by Number of Shows Directed')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
